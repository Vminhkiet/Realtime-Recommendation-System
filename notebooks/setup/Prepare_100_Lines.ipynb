```python
import os

# 1. C·∫•u h√¨nh ƒë∆∞·ªùng d·∫´n (Trong Docker)
base_path = "/home/jovyan/work/data_landing"

# File g·ªëc (N·∫∑ng)
input_file = f"{base_path}/otto/test.jsonl"

# File ƒë√≠ch (Nh·∫π - 100 d√≤ng)
output_file = f"{base_path}/otto_100.jsonl"

print(f"‚úÇÔ∏è ƒêang c·∫Øt 100 d√≤ng t·ª´ {input_file}...")

try:
    # M·ªü file g·ªëc (Read mode) v√† file ƒë√≠ch (Write mode)
    with open(input_file, 'r', encoding='utf-8') as f_in, \
         open(output_file, 'w', encoding='utf-8') as f_out:
        
        count = 0
        for line in f_in:
            if count >= 100: # ƒê·ªß 100 d√≤ng th√¨ d·ª´ng ngay
                break
            
            f_out.write(line)
            count += 1
            
    print(f"‚úÖ Xong! ƒê√£ t·∫°o file m·∫´u t·∫°i: {output_file}")
    print(f"üìä T·ªïng s·ªë d√≤ng: {count}")

except FileNotFoundError:
    print(f"‚ùå L·ªói: Kh√¥ng t√¨m th·∫•y file g·ªëc t·∫°i {input_file}. Ki·ªÉm tra l·∫°i ƒë∆∞·ªùng d·∫´n!")
```
