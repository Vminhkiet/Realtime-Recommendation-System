```python
import json
import time
import random
from kafka import KafkaProducer

# 1. K·∫øt n·ªëi Kafka (C·ªïng n·ªôi b·ªô Docker)
producer = KafkaProducer(
    bootstrap_servers=['kafka:29092'],
    value_serializer=lambda x: json.dumps(x).encode('utf-8')
)
```

```python
print("üöÄ ƒêang b∆°m d·ªØ li·ªáu Real-time (B·∫•m Stop ƒë·ªÉ d·ª´ng)...")

# Danh s√°ch s·∫£n ph·∫©m gi·∫£ l·∫≠p
products = [101, 102, 201, 301, 404]
actions = ['click', 'view', 'add_to_cart']
```

```python
# 2. V√≤ng l·∫∑p v√¥ t·∫≠n (Gi·∫£ l·∫≠p d√≤ng ch·∫£y)
try:
    while True:
        # T·∫°o random 1 s·ª± ki·ªán
        event = {
            "session": random.randint(1, 5), # Gi·∫£ l·∫≠p 5 user ƒëang online
            "aid": random.choice(products),
            "type": random.choice(actions),
            "ts": int(time.time() * 1000)
        }
        
        # G·ª≠i v√†o topic 'user_clicks'
        producer.send('user_clicks', value=event)
        print(f"‚û°Ô∏è G·ª≠i: User {event['session']} - {event['type']} - SP {event['aid']}")
        
        # Ch·ªù 1 ch√∫t (0.5 gi√¢y) ƒë·ªÉ d·ªÖ nh√¨n
        time.sleep(0.5)

except KeyboardInterrupt:
    print("üõë ƒê√£ d·ª´ng b∆°m.")
    producer.close()
```
